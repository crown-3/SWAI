```python
# package import하기
from bs4 import BeautifulSoup

# 문서 import하기 (같은 경로에 있는 경우)
with open("BS_Exercise(UTF-8).html", encoding="utf-8") as ex:
    soup = BeautifulSoup(ex, 'html.parser')
```

### 지시 사항 1
앞 페이지 코딩 결과로 soup 변수에 있는 html 코드 중에서 사진 부분의 정보를 모두 출력하고자 합니다.
결과는 아래 그림처럼 총 100 개가 출력되도록 빈칸의 코드를 완성하시오.

```python
img_src = soup.find('div', 'flex_grid credits search_results').find_all('img')
print(len(img_src))
```

### 지시 사항 2
for 반복문을 활용하여 아래와 같은 결과가 나오도록 코딩하시오.
100개의 img_src 목록에서 src 속성만 추출한 결과입니다.

```python
# src 속성만 추출합니다 (tag.attrs 이용)
img_src_srcs = src_list = [img_tags['src'] for img_tags in img_src if 'src' in img_tags.attrs]
```

```python
len(img_src_srcs)
```

```python
for src in img_src_srcs:
    print(src)
```

### 지시 사항 3, 4
추출된 결과 중에서 2017 / 2018년 데이터를 각각 img_src_2017 / img_src_2018 리스트에 저장하여 출력하는 코드를 작성하시오.

```python
# 배열을 선언하고, 해당 문자열이 포함된 태그들만 저장
img_src_2017, img_src_2018 = [], []

for src in img_src_srcs:
    if '/2017/' in src:
        img_src_2017.append(src)
    elif '/2018/' in src:
        img_src_2018.append(src)
```

```python
def print_2017():
    print(f"2017년 데이터는 총 {len(img_src_2017)}건입니다")
    for src in img_src_2017:
        print(src)
        
def print_2018():
    print(f"2018년 데이터는 총 {len(img_src_2018)}건입니다")
    for src in img_src_2018:
        print(src)
        
```

```python
print_2017()
```

```python
print_2018()
```
