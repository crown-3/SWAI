```python
import re
from konlpy.tag import *
okt = Okt()
kkma = Kkma()
```

```python
dbk_novel_data = open("./동백꽃.txt", encoding="utf-8").read()
print(dbk_novel_data)
```

```python
def remove_paragraph_numbers(text):
    # 숫자와 그 뒤의 줄바꿈 문자를 제거합니다.
    cleaned_text = re.sub(r"^\d+\n", "", text, flags=re.MULTILINE)
    return cleaned_text
    
dbk_novel_data = remove_paragraph_numbers(dbk_novel_data)

dbk_sentences =  kkma.sentences(dbk_novel_data)
dbk_sentences
```
